Function
protein, carb and all below function's functionality is similar
- First in recipe's coloumn (ingredients only) value are taken as input and split based on '\n'
- Search whether the  ingredient value is present in lst_ingredient,then takng its index and assigned to ind
- same index value of ingredient value i.e ind , is put to the list of protein present in lst_ingredient  added to pro
- this process continues for all ingredients and protein value is retured 

```python
def protein(x):
    ingredients = x.split('\n')
    pro = 0

    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            pro = pro +list(nut_ingredient['protein'])[ind]
    return round(pro,2)

def carb(x):
    ingredients = x.split('\n')
    carb = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            carb = carb+list(nut_ingredient['carbs'])[ind]
    return round(carb,2)

def fat(x):
    ingredients = x.split('\n')
    fat = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            fat = fat+list(nut_ingredient['fat'])[ind]
    return round(fat,2)

def fibre(x):
    ingredients = x.split('\n')
    fib = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            fib = fib+list(nut_ingredient['fibre'])[ind]
    return round(fib,2) 

def calcium(x):
    ingredients = x.split('\n')
    cal = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            cal = cal+list(nut_ingredient['calcium'])[ind]
    return round(cal,2) 

def iron(x):
    ingredients = x.split('\n')
    iron = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            iron = iron+list(nut_ingredient['iron'])[ind]
    return round(iron,2) 

def energy(x):
    ingredients = x.split('\n')
    energy = 0
    for ingredient in ingredients:
        if ingredient.lower() in list(nut_ingredient['ingredients']):
            ind=list(nut_ingredient['ingredients']).index(ingredient.lower())
            energy = energy+list(nut_ingredient['energy'])[ind]
    return round(energy,2) 
```

```python
recipe.loc[:,'protein']=recipe['ingredients only'].apply(protein)
recipe.loc[:,'carb']=recipe['ingredients only'].apply(carb)
recipe.loc[:,'fat']=recipe['ingredients only'].apply(fat)
recipe.loc[:,'fibre']=recipe['ingredients only'].apply(fibre)
recipe.loc[:,'calcium']=recipe['ingredients only'].apply(calcium)
recipe.loc[:,'iron']=recipe['ingredients only'].apply(iron)
recipe.loc[:,'energy']=recipe['ingredients only'].apply(energy)
```

The above cuisine_dct i.e dictionary was done to map state name of the respective recipe 
The below Function add_states fist search the name of the state present in 'name of the dish' if it is present returns 
the state name else it is searched in 'cuisine' coloumn it done so ,because in few of the rows state name present in the col
'name of the dish' (for ex: consinder a row where 'name of the dish' col is 'Karanataka style ... ' where we can map state name 
and in the col 'cuisine'- it might be 'South Indian Recipes' so i considered 'name of the dish' first  )

```python
def add_states(x):
    ingredients = recipe.loc[x,'name of the dish']
    name_dish_split = ingredients.split()
    for name in name_dish_split:
        lst=list(cuisine_dct.keys())
        lst_lower=[x.lower()for x in lst]
        if name.lower() in lst_lower:
            return(cuisine_dct[name])
    if recipe.loc[x,'cuisine'] in list(cuisine_dct.keys()):
        return(cuisine_dct[recipe.loc[x,'cuisine']])
   
```

```python
recipe.loc[:,'state']=recipe.index.map(add_states)
```
